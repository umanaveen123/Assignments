```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder 
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('book.csv')
df
```

# Pre-Processing
### As the data is not in transaction formation 
### We are using transaction Encoder

```python
df=pd.get_dummies(df)
df.head()
```

## # Apriori Algorithm 


```python
frequentitems=apriori(df,min_support=0.2,use_colnames=True)
```

```python
frequentitems
```

```python
model=association_rules(frequentitems,metric='lift',min_threshold=0.2)
```

```python
model
```

### An leverage value of 0 indicates independence. Range will be [-1 1]
### A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
model.sort_values('lift',ascending = False)[0:20]
```
