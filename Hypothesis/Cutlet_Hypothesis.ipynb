```python
import pandas as pd 
import numpy as np
from scipy import stats 
```

 Make two Hypothesis one contradicting to other H0 : Null Hypothesis H1 : Alternative Hypthosis

Decide a cut-off value Significance 5% alpha = 0.05 As it is a two-tailed test alpha/2 = 0.025

```python
cutlet = pd.read_csv("Cutlets.csv")
cutlet
```

```python
cutlet.rename(columns={'Unit A': 'Unit_A','Unit B':'Unit_B'})
```

```python
#2-sample 2-tail ttest:   stats.ttest_ind(array1,array2)    
statistic , p_value = stats.ttest_ind(cutlet.iloc[:,0],cutlet.iloc[:,1])
print(statistic , p_value)
```

```python
# In the above calculation ,P value is grater than Significance Value then we can accept Null Hypothesis
#there is no significant difference between the two sampless
```
